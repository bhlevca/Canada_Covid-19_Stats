These are time series plot of COVID 19 cases in all the provinces, territories and Canada. The goal is to see what the cases and deaths growth rate is, and whether actions by the federal and provincial governments are having an effect on the evolution of the pandemic in Canada.

__IMPORTANT NOTE (April 2022):__
It was and it still is my intention to update the page to provide important information related to the current phase of the `pandemic`, such as daily infections categorized by the vaccination status, as well as hospitalizations. This updated information would provide a better picture of the current epidemiological status and estimate if it transitioned to an endemic state.
However, Health Canada has declined to make public that data, which I requested since early December 2021. 

The official response to the request for the above requested data is as follows:

><em>Thank you for contacting the Public Health Agency of Canada (PHAC) to request information on the data set for Figure 6 in the COVID-19 daily epidemiology update. The national COVID-19 case data set upon which Figure 6 is based is not publicly available via a data repository. Detailed information on COVID-19 cases is reported to PHAC by the provinces and territories under a Multi-lateral Information Sharing Agreement. In order to protect the confidentiality of COVID-19 cases and under the terms of this agreement, we are unable to share more detailed information than is currently posted in our reports (COVID-19 daily epidemiology update, Weekly epidemiological report) and in the Preliminary dataset on confirmed cases of COVID-19.</em>

In my opinion the epidemiology reports based on vaccination status do not capture corectly efficiency of the vaccination as they create bidirectional bias due to the almagamation of data.
Also, due to the change in the methods of reporting the infections, the numbers may be misleading. 
  
__To visualize the graphs described below please scroll in the bottom panel:__
* Reverse ranking for the first 5 provinces for total cases, total deaths, new cases and newdeaths
* Total cases, total deaths, new cases and newdeaths for each of the provinces and territories and Canada
* Daily rate of growth starting February 2020
* Comparative cummulative cases in all provinces
* Prediction for the next 30 days in 4 most hit provinces based on the averaged growth rate from the last 5 days. This is a worst case scenario, for now/.
